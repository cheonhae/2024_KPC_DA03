```python
# 에어코리아 미세먼지 등급별 일수
import pandas as pd
import matplotlib.pyplot as plt
filepath = 'https://github.com/cheonhae/2024_KPC_DA_02/blob/master/4%E1%84%8C%E1%85%A1%E1%86%BC/04_data1.xlsx' +'?raw=true'
data = pd.read_excel(filepath, index_col='area')
data.head()
```

```python
# 2020년 미세먼지 좋음(2020_good) 데이터 가져오기
ata2020_good = data['2020_good']
data2020_good
```

```python
#2020년 미세먼지 좋음(2020_good) 데이터 그래프 그리기
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='b', marker='o')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 Fine Dust(pm10) Good Line Graph')
plt.grid()
plt.show()
```

```python
# 2015~ 2020년 미세먼지 좋음(2020_good) 데이터 가져온 후 그래프 그리기
plt.figure(figsize=(20,4))

for year in range(2015, 2021):
  year = str(year)+'_good'
  chartdata = data[year]
  plt.plot(chartdata, marker='o', label=year)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 세로 막대 그래프 그리기 : bar()
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel(filepath, index_col='area')
data2015_verybad = data['2015_verybad']

plt.figure(figsize=(20,4))
plt.bar(data2015_verybad.index, data2015_verybad, color='g')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015 Fine Dust(pm10) Very Bad Bar Graph')
plt.grid()
plt.show()
```

```python
# 그룹 세로 막대그래프 그리기: bar(), width
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))

data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range(2015, 2021):
  year = str(year)+'_good'
  chartdata = data6[year]
  plt.bar(index, chartdata, width=0.15, label=year) # width 를 변경해보자
  index = index + 0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
# 그룹 누적 가로 막대 그래프 그리기: barh()
import numpy as np
index = np.arange(7)

plt.figure(figsize=(10,4))

data6 = data.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017, 2021):
  year = str(year)+'_common'
  chartdata = data6[year]
  plt.???(index, chartdata, label=year)

plt.xlim(120,310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
# 산점도 그리기
import pandas as pd
import matplotlib.pyplot as plt
filepath = 'https://raw.githubusercontent.com/cheonhae/2024_KPC_DA_02/refs/heads/master/4%E1%84%8C%E1%85%A1%E1%86%BC/04_data2.csv'
data = pd.read_csv(filepath)
data
```

```python
# 산점도 그리기: scatter()
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

plt.figure(figsize=(10,4))
plt.????(PetalLength, PetalWidth)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()
```

```python
# 그룹 산점도 그리기: scatter()
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w', label='PetalLength * PetalWidth')
plt.scatter(SepalLength, SepalWidth, color='b', edgecolor='w', label='SepalLength * SepalWidth')

plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Petal Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 히스토그램 그리기 : hist()
setosa = data.loc[data.Species=='setosa',['Species','PetalLength']]

plt.figure(figsize=(10,6))
plt.???(setosa['PetalLength'], bins=8, label='Petal Length')

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Setosa Petal Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 그룹 히스토그램 그리기
versicolor_PetalLength = data.loc[data.Species=='versicolor',['Species','PetalLength']]
virginica_PetalLength = data.loc[data.Species=='virginica',['Species','PetalLength']]

plt.figure(figsize=(10,6))
plt.hist(versicolor_PetalLength['PetalLength'], bins=5, alpha=0.5, label='versicolor')
plt.hist(virginica_PetalLength['PetalLength'], bins=5, alpha=0.5, label='virginica')

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Petal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상자 수염(boxplot) 그리기
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(8,5))
plt.????([SepalLength, SepalWidth], labels=['Sepal Length','Sepal Width'])

plt.xlabel('Sepal')
plt.ylabel('Value')
plt.title('Iris Sepal Box Plot')
plt.show()
```

```python
# 가로 상자 수염(boxplot) 그리기: boxplot(), vert=False
plt.figure(figsize=(8,5))
plt.boxplot([SepalLength, SepalWidth], labels=['Sepal Length','Sepal Width'], vert=False)

plt.ylabel('Sepal')
plt.xlabel('Value')
plt.title('Iris Sepal Horizontal Box Plot')
plt.show()
```
